```python
all_files = os.listdir('Data')
```

```python
plt.figure()
plt.xlabel('Frequency [MHz]')
plt.ylabel('Amplification [Mag]')
for one_file in all_files:
    data = np.loadtxt(os.path.join('Data', one_file),skiprows=7,delimiter=',')
    plt.plot(data[:,0]/1e6,data[:,5])
```
